```python
##TAXA DE DESCONTO, TARIFA E CUSTOS
# fazendo 1000 x - 1 amostra de 30 elementos (uma taxa para cada ano da vida útil)
# usa o método Latin Hipercube Sampling para determinar os valores
#Obs: o arquivo LHS_amostra será criado pelo programa e salvo na pasta onde esse script estiver salvo.
#Obs: para que esse script rode, é necessário instalar o módulo pyDOE.

from pyDOE import lhs
 
    
####Taxadedesconto
#Gerando as amostras LHS:
lista_td=[] 
for l1 in range (1000): #repete 1000 vezes
    LHS1=lhs(30,1, None, None) #Serão geradas 1 amostra de 30 variáveis, cujos valores variam entre 0 e 1, dispostas em uma única coluna.
    ajuste1= (LHS1*0.9 + 0.7) #ajuste ao intervalo correto - nesse exemplo, o intervalo é 0.7 - 1.60. 
    lista_td.append(((((((str(ajuste1).replace('\n','')).replace(" ",",")).replace(",,",",")).replace(",,",",")).replace(",,",",")).replace("[[","")).replace("]]","")) #para retirar os espaços e \n da string gerada
   #Exportação das amostras LHS gerada para o documento LHS_amostra.csv
with open('Taxadedesconto.csv', 'w',encoding = 'UTF-8') as LHS_doc:
   for j1 in lista_td:
        LHS_doc.write((((j1.strip(','))+'\n'))) #\n para ficar em linhas separadas e tirar a vírgula do final
#Obs: o arquivo LHS_amostra será criado pelo programa e salvo na pasta onde esse script estiver salvo.
#Obs: para que esse script rode, é necessário instalar o módulo pyDOE.

####Tarifa
#Gerando as amostras LHS:
lista_tar=[] 
for l2 in range (1000): 
    LHS2=lhs(30,1,None,None)
    ajuste2= (LHS2* 0.25) + 1 #ajuste ao intervalo correto - nesse exemplo, o intervalo é 1 - 1.25. 
    lista_tar.append(((((((str(ajuste2).replace('\n','')).replace(" ",",")).replace(",,",",")).replace(",,",",")).replace(",,",",")).replace("[[","")).replace("]]","")) #para retirar os espaços e \n da string gerada
   #Exportação das amostras LHS gerada para o documento LHS_amostra.csv   
with open('Tarifa.csv', 'w',encoding = 'UTF-8') as LHS_doc:
   for j2 in lista_tar:
        LHS_doc.write((((j2.strip(','))+'\n')))
    
####Custos
#Gerando as amostras LHS:
lista_cust=[] 
for l3 in range (1000): 
    LHS3=lhs(30,1,None,None)
    ajuste3= (LHS3* 0.17) + 1.03 #ajuste ao intervalo correto - nesse exemplo, o intervalo é 1.03 - 1.20. 
    lista_cust.append(((((((str(ajuste3).replace('\n','')).replace(" ",",")).replace(",,",",")).replace(",,",",")).replace(",,",",")).replace("[[","")).replace("]]","")) #para retirar os espaços e \n da string gerada
   #Exportação das amostras LHS gerada para o documento LHS_amostra.csv   
with open('Custos.csv', 'w',encoding = 'UTF-8') as LHS_doc:
   for j3 in lista_cust:
        LHS_doc.write((((j3.strip(','))+'\n')))
```
