```python
## DISCOUNT RATE, TARIFF AND COSTS
# making - 1 sample of 1000 elements (same rate for all years of life span)
# uses the Latin Hipercube Sampling method to determine the values
#Obs: The files are created by the program and saved to the folder where that script is saved.
#Obs: For this script to run, you must install the pyDOE module (prompt - pip install pyDOE)

from pyDOE import lhs
import numpy as np
import csv

####DISCOUNT RATE
# Generating LHS samples:
list_dr=[] 
LHS1=lhs(1000,1, None, None) # 1 sample of 1000 variables, whose values vary between 0 and 1, will be generated, arranged in a single column.
adjust1= (LHS1*0.9 + 0.7) # adjust to the correct range - in this example, the range is 1 - 1.60.
adjust1T=adjust1.T
list_dr=(adjust1T.tolist())
# Export of LHS samples generated to csv
with open('Discountrate_cte.csv', 'w',encoding = 'UTF-8',newline='') as LHS_doc:
    for j1 in list_dr:
        n=1
        while n<30:
            LHS_doc.write(((str(j1).replace('[','')).replace("]",""))+',')
            n+=1
        LHS_doc.write(((str(j1).replace('[','')).replace("]",""))+'\n')  

####TARIFF
# Generating LHS samples:
list_tar=[] 
LHS2=lhs(1000,1, None, None) # 1 sample of 1000 variables, whose values vary between 0 and 1, will be generated, arranged in a single column.
adjust2= (LHS2*0.25)+1 # adjust to the correct range - in this example, the range is 1 - 1.25.
adjust2T=adjust2.T
list_tar=(adjust2T.tolist())
# Export of LHS samples generated to csv
with open('Tariff_cte.csv', 'w',encoding = 'UTF-8',newline='') as LHS_doc:
    for j2 in list_tar:
        n=1
        while n<30:
            LHS_doc.write(((str(j2).replace('[','')).replace("]",""))+',')
            n+=1
        LHS_doc.write(((str(j2).replace('[','')).replace("]",""))+'\n')
        
####COSTS
# Generating LHS samples:
list_cost=[] 
LHS3=lhs(1000,1, None, None) # 1 sample of 1000 variables, whose values vary between 0 and 1, will be generated, arranged in a single column.
adjust3= (LHS3*0.17)+1.03 # adjust to the correct range - in this example, the range is 1.00 - 1.20.
adjust3T=adjust3.T
list_cost=(adjust3T.tolist())
# Export of LHS samples generated to csv
with open('Costs_cte.csv', 'w',encoding = 'UTF-8',newline='') as LHS_doc:
    for j3 in list_cost:
        n=1
        while n<30:
            LHS_doc.write(((str(j3).replace('[','')).replace("]",""))+',')
            n+=1
        LHS_doc.write(((str(j3).replace('[','')).replace("]",""))+'\n')  
```
